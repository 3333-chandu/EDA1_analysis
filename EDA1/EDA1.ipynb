```python
columns=data.columns
columns
```

```python
# calculating mean, mode ,median ,std
for column in columns:
    print(f"\nColumn name: {column}")
    print("Mean:", data[column].mean())
    print("Median:", data[column].median())
    print("Mode:", data[column].mode()[0])
    print("Standard Deviation:", data[column].std())
```

```python
#finding the ouliers 
import pandas as pd

# Let's assume 'data' is your DataFrame and 'numerical_columns' contains numeric column names
Q1 = data[columns].quantile(0.25)
Q3 = data[columns].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers: A value is an outlier if it's below (Q1 - 1.5*IQR) or above (Q3 + 1.5*IQR)
outlier_condition = (data[columns] < (Q1 - 1.5 * IQR)) | (data[columns] > (Q3 + 1.5 * IQR))

# Print number of outliers per column
print("Outliers in each column:")
print(outlier_condition.sum())
```

```python
#step3:data visualization
#creating histogram model for numerical columns
import matplotlib.pyplot as plt
import seaborn as sns

for column in columns:
    plt.figure(figsize=(8,4))
    sns.histplot(data[column], kde=True)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
    
```

```python
#boxplot for numerical columns
for col in columns:
    plt.figure()
    data.boxplot(column=col)
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.show()
    print()
```

```python
# finding any categorical data present in it
categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical columns:", categorical_columns)
```

```python
# here there is no categorical columns present in the data
#we can't make a barchart here
# the below code defines how to create a barchart if there is  categorical columns present
for col in categorical_columns:
    print(f"Bar chart for: {col}")
    sns.countplot(x=col, data=data)
    plt.title(f'{col} Distribution')
    plt.show()

    print(f"Pie chart for: {col}")
    data[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f'{col} Distribution')
    plt.ylabel("")
    plt.show()
```

```python
corelation=data.corr(numeric_only=True)
```

```python
corelation
```

```python
#	Generating correlation heatmaps 
##In heatmap wherever the corelation is high ,the colour is darker shade in above map.

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))  
sns.heatmap(
    corelation.corr(),
    annot=True, 
    xticklabels=corelation.columns,
    yticklabels=corelation.columns,      
    fmt=".2f",                
    cmap="coolwarm",          
    linewidths=0.5,           
    linecolor='gray',         
    square=True,
)
plt.title("Correlation Heatmap of Numerical Columns")
plt.tight_layout()
plt.show()
```

```python
#Strong positive correlations (close to +1):
# means two variables increase together 
#Strong negative correlations (close to -1):
# means when one variable goes up, the other goes down.

#Near 0 correlation:
# no linear relationship.
```
